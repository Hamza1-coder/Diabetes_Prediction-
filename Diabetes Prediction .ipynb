## Cumulative Distribution Function(CDF)

```python
count1, bin_edges1 = np.histogram(status_yes['Outcome'],bins=10, density= True)
pdf1 = count1/sum(count1)
print(pdf1)
print(bin_edges1)
cdf1 = np.cumsum(pdf1)
plt.plot(bin_edges1[1:], pdf1)
plt.plot(bin_edges1[1:], cdf1, label = 'Yes')
plt.xlabel('outcome')
print("*"*75)
count2, bin_edges2 = np.histogram(status_no['Outcome'],bins=10, density= True)
pdf2 = count2/sum(count2)
print(pdf2)
print(bin_edges2)
cdf2 = np.cumsum(pdf2)
plt.plot(bin_edges2[1:], pdf2)
plt.plot(bin_edges2[1:], cdf2, label = 'no')
plt.xlabel('outcome')

plt.legend()
plt.show()
```

```python
from sklearn.model_selection import train_test_split
X = df.drop(columns=['Outcome'])
y = df['Outcome']
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.03, random_state=10)
```

```python
from sklearn.impute import SimpleImputer
fill_values = SimpleImputer(missing_values=0, strategy='mean')
X_train = fill_values.fit_transform(X_train)
X_test = fill_values.fit_transform(X_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
rfm = RandomForestClassifier(random_state=10)
rfm.fit(X_train,y_train.ravel())
```

```python
predict_ = rfm.predict(X_test)
from sklearn import metrics
print("Accuracy = {0:.3f}". format(metrics.accuracy_score(y_test,predict_)))
```
